## Streaming data from a volume and writestream it into delta

```python
order_schema = StructType([
    StructField("order_id",       StringType(),  True),
    StructField("order_date",     StringType(),  True),
    StructField("ship_date",      StringType(),  True),
    StructField("delivery_date",  StringType(),  True),
    StructField("order_status",   StringType(),  True),
    StructField("customer_id",    StringType(),  True),
    StructField("customer_name",  StringType(),  True),
    StructField("country",        StringType(),  True),
    StructField("state",          StringType(),  True),
    StructField("city",           StringType(),  True),
    StructField("product_id",     StringType(),  True),
    StructField("product_name",   StringType(),  True),
    StructField("category",       StringType(),  True),
    StructField("sub_category",   StringType(),  True),
    StructField("brand",          StringType(),  True),
    StructField("quantity",       IntegerType(), True),
    StructField("unit_price",     DoubleType(),  True),
    StructField("discount",       DoubleType(),  True),
    StructField("shipping_cost",  DoubleType(),  True),
    StructField("total_sales",    DoubleType(),  True),
    StructField("payment_method", StringType(),  True)
])

df = spark.readStream.format("csv") \
    .schema(order_schema) .option("header", "true").option("inferSchema", True) \
    .option("multiline", "true") \
    .load("/Volumes/amazon_sales/bronze/streaming")
```

```python

df.writeStream.format("delta").outputMode("append").trigger(once=True).option("path", "/Volumes/amazon_sales/bronze/streaming/Data").option("checkpointLocation", "/Volumes/amazon_sales/bronze/streaming/jsonsink/checkpoint").start()
```

```python
#reading the delta table 
bronze_df = spark.read.format("delta").load("/Volumes/amazon_sales/bronze/streaming/Data/")
bronze_df.display()
```

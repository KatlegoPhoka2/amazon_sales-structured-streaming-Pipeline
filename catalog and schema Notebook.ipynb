##Creating catalog and schema in Databricks


```python
# Create Catalog if not exists

spark.sql(f"CREATE CATALOG IF NOT EXISTS amazon_sales")
```

```python
# Create schema if not exists

spark.sql(f"CREATE SCHEMA IF NOT EXISTS amazon_sales.bronze")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS  amazon_sales.silver")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS amazon_sales.gold")
```
